```python
from pathlib import Path

import pandas as pd
```

```python
data_dir = Path("../../data")
```

```python
ts_df = pd.read_csv(data_dir / "raw" / "filtered_ts.csv")
mirrors_products_df = pd.read_csv(data_dir / "raw" / "mirrors_products.csv")
```

```python
mirrors_mask = ts_df["NM_ID"].isin(mirrors_products_df["article"])
```

```python
mirrors_ts_df = ts_df[mirrors_mask].reset_index(drop=True)
```

```python
mirrors_ts_df.to_csv(data_dir / "processed" / "mirrors_ts.csv", index=False)
```
