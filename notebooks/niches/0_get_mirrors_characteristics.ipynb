```python
import sys
from pathlib import Path

import pandas as pd

sys.path.append("../..")

from src.utilities import get_float_from_string
```

```python
data_dir = Path("../../data")
```

```python
mirrors_products_df = pd.read_csv(data_dir / "raw" / "mirrors_products.csv")
```

```python
mirrors_products_characteristics_df = pd.DataFrame()

copy_columns = ["article", "subj_name", "subj_root_name"]
mirrors_products_characteristics_df[copy_columns] = mirrors_products_df[copy_columns]

numerical_product_parameters_columns = [
    "options_Габариты__Высота предмета",
    "options_Габариты__Ширина предмета",
    "options_Габариты__Вес без упаковки (кг)",
]

mirrors_products_characteristics_df[numerical_product_parameters_columns] = mirrors_products_df[
    numerical_product_parameters_columns
].map(get_float_from_string)

mirrors_products_characteristics_df.head()
```

```python
mirrors_products_characteristics_df.to_csv(
    data_dir / "processed" / "mirrors_products_characteristics.csv", index=False
)
```
